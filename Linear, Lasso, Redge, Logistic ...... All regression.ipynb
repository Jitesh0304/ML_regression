```python
 # boston house pricing data
from sklearn .datasets import load_boston
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
dataset
```

```python
dataset['Price'] = df.target
dataset
```

```python
# Dividing the dataset into independent and dependent features

x = dataset.iloc[:,:-1]    #  x is the independent feature      
y = dataset.iloc[:,-1]     #  y is the dependent feature   
```

```python
x
```

```python
y
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x , y, test_size = 0.3 , random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_regr = LinearRegression()

mse = cross_val_score(lin_regr,x,y,scoring = 'neg_mean_squared_error' , cv = 5)

print(mse)
```

```python
mean_mse = np.mean(mse)              # mean value of those 5 value
print(mean_mse)
```

```python
mse = cross_val_score(lin_regr,x_train,y_train,scoring = 'neg_mean_squared_error' , cv = 5)

mean_mse = np.mean(mse)              # mean value of those 5 value
print(mean_mse)
```

```python
lin_regr.fit(x_train,y_train)
```

```python
y_pred1 = lin_regr.predict(x_test)
y_pred1
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df_c = load_breast_cancer()
```

```python
df_c
```

```python
X = pd.DataFrame(df_c['data'], columns = df_c['feature_names'])
```

```python
X
```

```python
y = pd.DataFrame(df_c['target'] , columns = ['Target'])
```

```python
y
```

```python
y['Target'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X , y, test_size = 0.3 , random_state = 42)
```

```python
params = [{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1 = LogisticRegression(C=10 ,max_iter=100)
```

```python
model = GridSearchCV(model1, param_grid = params , scoring ='f1' , cv = 5)
```

```python
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred4 = model.predict(x_test)
```

```python
y_pred4
```

## confussion matrix

```python
confusion_matrix(y_test,y_pred4)
```
