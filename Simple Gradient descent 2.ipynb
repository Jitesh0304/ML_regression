```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):                                    
    return x**2 + x + 1    
def df(x):                                    
    return 2*x + 1 
x_1 = np.linspace(-4,3,200)
plt.figure(figsize = (15,6))

plt.subplot(1,2,1)                                          
plt.plot (x_1 , f(x_1) , c = 'black' , linewidth= 3)                           
plt.grid()

plt.subplot(1,2,2)                                                              

plt.plot(x_1 , df(x_1) , linewidth = 3 , c = "red")
plt.grid()
plt.show()
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.1, precision=0.001, max_iter=200):
    
    new_x = initial_guess
    
    x_list = [new_x]
    
    slope_list = [derivative_func(new_x)]
    
    
    for i in range(max_iter):
        
        
        previous_x = new_x
        
        gradient = derivative_func(previous_x)               # slop
        
        new_x = previous_x - multiplier * gradient
        
        step_size = abs(new_x - previous_x)
        
        x_list.append(new_x)
        
        slope_list.append(derivative_func(new_x))
        
        if step_size < precision:
            
            break 
            
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func = df , initial_guess = 3, precision=0.00001)
```

```python
plt.figure(figsize =(18,8))


plt.subplot(1,2,1)

plt.grid()
plt.plot(x_1, f(x_1))

values = np.array(list_x)

plt.scatter(list_x, f(values), color = 'red')
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.subplot(1,2,2)

plt.grid()
plt.plot(x_1, df(x_1))
plt.scatter(list_x, deriv_list, color = 'navy')
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)


plt.show()
```
