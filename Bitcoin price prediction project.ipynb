```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
data_1 = pd.read_csv('bitcoin.csv')
```

```python
data_1.drop(['Date'],1,inplace=True)        
```

```python
predictionday = 30                                                      

data_1['Prediction'] = data_1[['Price']].shift(predictionday)
data_1.head()
```

```python
predictionday = 30                                                        

data_1['Prediction'] = data_1['Price'].shift(predictionday)

data_1.head()
```

```python
predictionday = 30                                                          
                                                                                    
data_1['Prediction'] = data_1['Price'].shift(-predictionday)
data_1.head()
```

```python
data_1.tail()
```

```python
data_1.shape
```

```python
x = np.array(data_1.drop(['Prediction'], axis = 1))
x = x[:len(data_1)-predictionday]

print(x)
```

```python
len(x)
```

```python
# Create the dependent data set
# convert the data frame into a numpy array

y = np.array(data_1['Prediction'])

y = y[:-predictionday]

print(y)
```

```python
len(y)
```

```python
## train test split

xtrain , xtest , ytrain , ytest = train_test_split(x , y , test_size = 0.2)
```

```python
# set the predictiondays array equal to all rows from the original data set                                                                                         
predictionday_array = np.array(data_1.drop(['Prediction'] , axis = 1))
print(predictionday_array)
```

```python
# set the predictiondays array equal to last 30 rows from the original data set

predictionday_array = np.array(data_1.drop(['Prediction'] , axis = 1))[-predictionday:]
print(predictionday_array)
```

```python
svr_rbf = SVR(kernel = 'rbf' , C = 1e3 , gamma = 0.00001)
svr_rbf.fit(xtrain , ytrain)
```

```python
# testing

svr_rbf_confidence = svr_rbf.score(xtest , ytest)

print('SVR_RBF_ accuracy :', svr_rbf_confidence)
```

```python
# print the prediction values

svr_prediction = svr_rbf.predict(xtest)
print(svr_prediction)

print()

print(ytest)
```

```python
# Print the model predictions for the next 30 days

svr_prediction = svr_rbf.predict(predictionday_array)
print(svr_prediction)
print()

#Print the actual price for bitcoin for last 30 days

print(data_1.tail(predictionday))
```
