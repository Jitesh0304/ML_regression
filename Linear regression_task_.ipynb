```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

### getting data

```python
data = pd.read_csv('lsd_math_score_data.csv')
data
```

```python
type(data)
```

```python
time = data[['Time_delay']]
drug = data[['Drug_ppm']]
score = data[['Math_score']]
```

```python
time
```

```python
drug
```

```python
score
```

```python
plt.figure(figsize = (15,6))
plt.title('SCATTER PLOT OF LSD vs MATHS SCORE', fontsize = 25)
plt.xlabel('LSD in PPM in THE BLOOD',fontsize = 20 )
plt.ylabel('MATHS SCORE', fontsize = 20)
plt.text(x=5.1, y = 74, s = 'DSP 28 PLOT', fontsize = 20) 
plt.scatter(drug, score, color = 'red', s=100, alpha = 0.9) 
plt.show()
```

### for best fit line drawing

```python
regr = LinearRegression()
```

```python
regr.fit(drug , score)    
```

```python
regr.intercept_   
```

```python
regr.coef_      
```

```python
# y = mx + c                                                             
prediction = regr.coef_ * drug + regr.intercept_                         
prediction.values                                                        
```

```python
plt.figure(figsize = (15,6))
plt.title('REGRESSION and SCATTER PLOT OF LSD vs MATHS SCORE', fontsize = 25)
plt.xlabel('LSD in PPM in THE BLOOD',fontsize = 20 )
plt.ylabel('MATHS SCORE', fontsize = 20)
plt.text(x=5.1, y = 74, s = 'DSP 28 PLOT', fontsize = 20)
plt.scatter(drug, score, color = 'red', s=100, alpha = 0.9)                 
plt.plot(drug.values, prediction.values, color = 'navy', alpha = 0.9)        
plt.show()
```
